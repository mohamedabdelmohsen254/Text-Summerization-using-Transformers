# Import and Load Model

**PegasusForConditionalGeneration** model is a pre-trained model that has been trained on large amounts of text data. It is specifically designed for text generation tasks, such as summarization, where you want to generate a concise summary of a longer text.

**PegasusTokenizer** is a tool that helps with tokenization, which is the process of breaking down a text into smaller units, such as words or subwords. Tokenization is an important step in NLP tasks as it helps in processing and understanding the text.

```python
import torch #if you get error here, install pytorch

from transformers import PegasusForConditionalGeneration, PegasusTokenizer

from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
```

**from_pretrained** method is a convenient way to load a pre-trained tokenizer provided by the Hugging Face's Transformers library. In this case, it loads the PegasusTokenizer specifically designed for the "google/pegasus-xsum" model.

```python
# Load tokenizer 
tokenizer = PegasusTokenizer.from_pretrained("google/pegasus-xsum")
```

```python
# Load model 
model = PegasusForConditionalGeneration.from_pretrained("google/pegasus-xsum")
```

# Perform Abstractive Summarization

```python
text = """
Python is an interpreted high-level general-purpose programming language. Its design philosophy emphasizes code readability with its use of significant indentation. Its language constructs as well as its object-oriented approach aim to help programmers write clear, logical code for small and large-scale projects.[30]

Python is dynamically-typed and garbage-collected. It supports multiple programming paradigms, including structured (particularly, procedural), object-oriented and functional programming. It is often described as a "batteries included" language due to its comprehensive standard library.[31]

Guido van Rossum began working on Python in the late 1980s, as a successor to the ABC programming language, and first released it in 1991 as Python 0.9.0.[32] Python 2.0 was released in 2000 and introduced new features, such as list comprehensions and a garbage collection system using reference counting. Python 3.0 was released in 2008 and was a major revision of the language that is not completely backward-compatible. Python 2 was discontinued with version 2.7.18 in 2020.[33]

Python consistently ranks as one of the most popular programming languages.[34][35][36][37]"""
```

**Tokenization:** The tokenizer object, which was previously loaded, is used to tokenize the input text. The input text is passed as the text parameter.

**Truncation:** The truncation=True argument specifies that if the input text is longer than the maximum allowed length, it should be truncated to fit the model's maximum input length. This ensures that the input is within the model's capacity. The "google/pegasus-xsum" variant, has a maximum sequence length of 1024 tokens.

**Padding:** The padding="longest" argument indicates that if the input text is shorter than the maximum allowed length, it should be padded with special tokens or zeros to match the maximum length. Padding ensures that all input sequences have the same length, which is necessary for batch processing.

**Return Tensors:** The return_tensors="pt" argument specifies that the tokens should be returned as PyTorch tensors. This facilitates compatibility with the Pegasus model, which typically expects input in tensor format.

```python
# Create tokens - number representation of our text
tokens = tokenizer(text, truncation=True, padding="longest", return_tensors="pt")
```

```python
tokens
```

By calling model.generate(**tokens), you invoke the Pegasus model's capabilities to generate a summary based on the provided tokenized input. 

```python
# Summarize 
summary = model.generate(**tokens)
```

```python
summary[0]
```

```python
# Decode summary
generated_summary = tokenizer.decode(summary[0], skip_special_tokens=True)
generated_summary
```

```python
reference_text = text.strip()  # Remove leading/trailing white spaces
reference_text
```

```python
from rouge_score import rouge_scorer

scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'])
scores = scorer.score(generated_summary, reference_text)
```

```python
# Step 4: Access ROUGE Scores
rouge1_precision = scores['rouge1'].precision
rouge1_recall = scores['rouge1'].recall
rouge1_fmeasure = scores['rouge1'].fmeasure

rouge2_precision = scores['rouge2'].precision
rouge2_recall = scores['rouge2'].recall
rouge2_fmeasure = scores['rouge2'].fmeasure

rougeL_precision = scores['rougeL'].precision
rougeL_recall = scores['rougeL'].recall
rougeL_fmeasure = scores['rougeL'].fmeasure


# Print ROUGE Scores
print("ROUGE-1 Precision:", rouge1_precision)
print("ROUGE-1 Recall:", rouge1_recall)
print("ROUGE-1 F-measure:", rouge1_fmeasure)
print()
print("ROUGE-2 Precision:", rouge2_precision)
print("ROUGE-2 Recall:", rouge2_recall)
print("ROUGE-2 F-measure:", rouge2_fmeasure)
print()
print("ROUGE-L Precision:", rougeL_precision)
print("ROUGE-L Recall:", rougeL_recall)
print("ROUGE-L F-measure:", rougeL_fmeasure)
```

In general, the ROUGE scores indicate that the generated summary has relatively low precision, recall, and F-measure values across all the evaluated metrics. This suggests that the generated summary does not match the reference text well in terms of unigrams, bigrams, and longest common subsequences. Further improvements may be needed to enhance the quality and content overlap between the generated summary and the reference text.

# Try Again but with Preprocessing

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import string


# Text preprocessing steps
def preprocess_text(text):
    # Lowercasing
    text = text.lower()
    
    # Removing punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    
    # Tokenization
    tokens = word_tokenize(text)
    
    # Removing stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [token for token in tokens if token not in stop_words]
    
    # Lemmatization
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(token) for token in tokens]
    
    # Join tokens back into a string
    preprocessed_text = ' '.join(tokens)
    
    return preprocessed_text

# Example usage
text = """
Python is an interpreted high-level general-purpose programming language. Its design philosophy emphasizes code readability with its use of significant indentation. Its language constructs as well as its object-oriented approach aim to help programmers write clear, logical code for small and large-scale projects.[30]

Python is dynamically-typed and garbage-collected. It supports multiple programming paradigms, including structured (particularly, procedural), object-oriented and functional programming. It is often described as a "batteries included" language due to its comprehensive standard library.[31]

Guido van Rossum began working on Python in the late 1980s, as a successor to the ABC programming language, and first released it in 1991 as Python 0.9.0.[32] Python 2.0 was released in 2000 and introduced new features, such as list comprehensions and a garbage collection system using reference counting. Python 3.0 was released in 2008 and was a major revision of the language that is not completely backward-compatible. Python 2 was discontinued with version 2.7.18 in 2020.[33]

Python consistently ranks as one of the most popular programming languages.[34][35][36][37]"""

preprocessed_text = preprocess_text(text)
print(preprocessed_text)
```

```python
# Create tokens - number representation of our text
tokens = tokenizer(preprocessed_text, truncation=True, padding="longest", return_tensors="pt")
tokens
```

```python
# Summarize 
summary = model.generate(**tokens)
```

```python
# Decode summary
generated_summary = tokenizer.decode(summary[0], skip_special_tokens=True)
generated_summary
```

```python
reference_text = preprocessed_text.strip()  # Remove leading/trailing white spaces
reference_text
```

```python
from rouge_score import rouge_scorer

scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'])
scores = scorer.score(generated_summary, reference_text)
```

```python
# Step 4: Access ROUGE Scores
rouge1_precision = scores['rouge1'].precision
rouge1_recall = scores['rouge1'].recall
rouge1_fmeasure = scores['rouge1'].fmeasure

rouge2_precision = scores['rouge2'].precision
rouge2_recall = scores['rouge2'].recall
rouge2_fmeasure = scores['rouge2'].fmeasure

rougeL_precision = scores['rougeL'].precision
rougeL_recall = scores['rougeL'].recall
rougeL_fmeasure = scores['rougeL'].fmeasure


# Print ROUGE Scores
print("ROUGE-1 Precision:", rouge1_precision)
print("ROUGE-1 Recall:", rouge1_recall)
print("ROUGE-1 F-measure:", rouge1_fmeasure)
print()
print("ROUGE-2 Precision:", rouge2_precision)
print("ROUGE-2 Recall:", rouge2_recall)
print("ROUGE-2 F-measure:", rouge2_fmeasure)
print()
print("ROUGE-L Precision:", rougeL_precision)
print("ROUGE-L Recall:", rougeL_recall)
print("ROUGE-L F-measure:", rougeL_fmeasure)
```
